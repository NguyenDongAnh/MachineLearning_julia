```julia
X = [ones(length(y)) A[:,2:end]]
```

```julia
k=unique(y)
k=sort(k)
```

```julia
(N, D) = size(X)
```

```julia
δ = zeros(N,length(k))
for i = 1:length(k)
    δ[:,i] = (y .== i)  
end
δ
```

```julia
function loss(X,y,θ)::Float64
    u = sum(θ[y,:] .* X, dims=2)
    v = sum(exp.(θ*X'),dims=1)
    J = -sum(u)+sum(log.(v))
    J = J/N + 0.
end
```

```julia
function gradient(X, δ, θ)::Array{Float64}
    u=exp.(θ*X')
    z=sum(u,dims=1)
    P = u' ./ z'
    (P-δ)'*X/N
end
```

```julia
function hessian(X,θ)
    u=exp.(θ*X')
    z=sum(u,dims=1)
    P = u' ./ z'
    a = map(i->Diagonal(P[:,i]),1:length(k))
    b = map(t->P[:,t]*P[:,t]',1:length(k))
    map(t->((X'*(a.-b)[t]*X))^-1,1:length(k))
end
```

```julia
function bgd(X,δ,α,λ, maxIterations=100)
    (N, D) = size(X)
    θ = zeros(length(k),D)
    θs = []
    push!(θs, θ)
    for i = 1:maxIterations
        ∇ = gradient(X,δ,θ)
        θ = θ - α*(∇)
#         θ = θ - α*(∇ + λ*[zeros(length(unique(y))) θ[:,2:end]])
        push!(θs, θ)
    end
    θs
end
```

```julia
I = zeros(length(k),D)
function bgd2(X,δ,α,λ, maxIterations=100)
    (N, D) = size(X)
    θ = zeros(length(k),D)
    θs = []
    push!(θs, θ)
    for i = 1:maxIterations
        ∇ = gradient(X,δ,θ)
        H = hessian(X,θ)
        for i = 1:length(k)
            I[i,:] = H[i]*∇[i,:]
        end
        θ = θ - α*I
#         θ = θ - α*(∇ + λ*[zeros(length(unique(y))) θ[:,2:end]])
        push!(θs, θ)
    end
    θs
end
```

```julia
θ_0 = zeros(D)
T = 10000
λ=0.1
α=0.01
θs = bgd2(X,δ,α,λ,T)
```

```julia
Js = map(θ -> loss(X, y, θ), θs)#lost ung voi θ
```

```julia
plot(1:T+1, Js,label = "loss", xlabel = "Itr",ylabel = "loss")
```

```julia
function classify(X, θ)
    h = θ*X
    argmax(h)
end
```

```julia
using Optim
```

```julia
θ_0 = zeros(length(k),D)
```

```julia
function cost(X,y,θ)::Float64
    u = sum(θ[y,:] .* X, dims=2)
    v = sum(exp.(θ*X'),dims=1)
    J = -sum(u)+sum(log.(v))
    J = J/N + 0.
end
```

```julia
function grad!(∇, θ)
    u=exp.(θ*X')
    z=sum(u,dims=1)
    P = exp.(X*θ') ./ z'
    temp = (P'-δ')*X/N
    n,d = size(temp)
#     r = λ*[zeros(length(k)) θ[:,2:end]]
    for i=1:n
        ∇[i,:]=temp[i,:] 
#         + r[i,:]
    end
end
```

```julia
result_neldermead = optimize(θ->cost(X,y,θ),grad!,θ_0,NelderMead())
```

```julia
θ_best_neldermead = Optim.minimizer(result_neldermead)
```

```julia
prediction = map(i->classify(X[i,:], θ_best_neldermead), 1:N)
prediction = prediction
accuracy = sum(y .== prediction)/N
```

```julia
cost(X,y,θ_best_neldermead)
```

```julia
result_lbfgs = optimize(θ->cost(X,y,θ),grad!,θ_0,LBFGS())
```

```julia
θ_best_lbfgs=Optim.minimizer(result_lbfgs)
```

```julia
prediction = map(i->classify(X[i,:], θ_best_lbfgs), 1:N)
accuracy = sum(y .== prediction)/N
```
