```julia
using DelimitedFiles
```

```julia
A = readdlm("D:/julia-workspace/wdbc.txt", ',', Float64, '\n')
```

```julia
y = A[:,2].+1
```

```julia
X = [ones(length(y)) A[:,3:31]]
```

```julia
(N,D) = size(X)
θ = zeros(D)
```

```julia
function loss(X::Array{Float64,2}, y::Array{Float64}, θ::Array{Float64})::Float64
    N = length(y)
    h = sigmoid.(X*θ)
    l = y'*log.(h)+(1 .- y)'*log.(1 .- h)
    J = -l/N + 0.
    J
end
```

```julia
function sigmoid(z)
    h = 1/(1+exp(-z))
end
```

```julia
function gradient(X, y, θ)::Array{Float64}
    N = length(y)
    h = sigmoid.(X*θ)
    X'*(h - y)/N .+ 0.

end
```

```julia
function bgd(X, y, θ_0, α = 0.001, T = 100)::Array{Array{Float64}}
    θ = θ_0
    θs = []
    for t=1:T
        ∇ = gradient(X, y, θ)
        θ = θ - α * ∇
        push!(θs, θ) # Day tung phan tu θ vao duoi
    end
    θs
end
```

```julia
θ_0 = zeros(D)
α = 0.01
T = 1000
θs = bgd(X, y, θ_0, α, T)
```

```julia
Js = map(θ -> loss(X, y, θ), θs)
θs[end,end]
```

```julia
function classify(X, θ)
    h = X'*θ
    if (h<0) 1.0 else 2.0 end
end
```

```julia
prediction = map(i->classify(X[i,:], θs[end,end]), 1:N)
accuracy = sum(y .== prediction)/N
```
