```julia
using Flux
```

```julia
X0 = Flux.Data.Iris.features()
```

```julia
y = Flux.Data.Iris.labels()
```

```julia
X = convert(Array{Float64,2},transpose(X0))
```

```julia
using Statistics
```

```julia
mean(X, dims=1)
```

```julia
x = [1 2 3]

repeat(x ,3 ,1)
```

```julia
function zero_mean(X::Array{Float64,2})::Array{Float64,2}
    μ = mean(X, dims=1)
    M = repeat(μ,size(X,1),1)
    X - M
end
```

```julia
X̄ = zero_mean(X)
```

```julia
mean(X̄,dims=1)
```

```julia
feature_order = [1, 3, 2, 4]
```

```julia
X̄ = zero_mean(X[:,feature_order])
```

```julia
using Plots
```

```julia
plot(X̄[:,1], X̄[:,2] ,st=:scatter)
```

```julia
labels = unique(y)
```

```julia
setosa = (y .== labels[1])
versicolor = (y .== labels[2])
virginica = (y .== labels[3])
```

```julia
X̄[setosa,2]
```

```julia
plot(X̄[setosa,1], X̄[setosa,2] ,st=:scatter, m=:+, c=:black, legend = false)
```

```julia
plot!(X̄[versicolor,1], X̄[versicolor,2] ,st=:scatter, m=:*, c=:red, legend = false)
```

```julia
plot!(X̄[virginica,1], X̄[virginica,2] ,st=:scatter, m=:o, c=:yellow, legend = false, xlabel = "Sepal Length", ylabel = "Petal Length")
```

```julia
cov(X̄)
```

```julia
S = (X̄'*X̄)/(length(y)-1)
```

```julia
using LinearAlgebra
```

```julia
#tinh vector rieng va gia tri rieng
values, vectors = eigen(S)
```

```julia
order = sortperm(values, rev = true)
```

```julia
θ = vectors[:,order]
```

```julia
Z = X̄*θ[:,1:2]
```

```julia
plot(Z[setosa,1], Z[setosa,2] ,st=:scatter, m=:+, c=:black, legend = false)
plot!(Z[versicolor,1], Z[versicolor,2] ,st=:scatter, m=:*, c=:red, legend = false)
plot!(Z[virginica,1], Z[virginica,2] ,st=:scatter, m=:o, c=:yellow, legend = false, xlabel = "First Component", ylabel = "Second Component")
```

```julia
cov(Z)
```
